```python
# Example usage
hot_vector = [8, 9, 7, 6, 10]
crazy_vector = [5, 6, 4, 7, 8]
area = plot_crazy_hot_matrix(hot_vector, crazy_vector, return_area=True, plot=True)
print("Center is located in:", area)
```

# Realistic Scenario

- A man may meet 10,000 women in his life.
- A man will only meet each woman no more than five times before he decides where that woman is located on the hot crazy matrix.

This is a low-information decision process in which most men find themselves.

```python
observations = []
for i in range(10000):
    n = 5
    hot_vector = np.random.normal(loc=5, scale=3, size = n )
    crazy_vector = np.random.normal(loc=6, scale=2, size = n )
    
    area = plot_crazy_hot_matrix(hot_vector, crazy_vector, return_area=True, plot=False)
    observations.append(area)
    print(i, end='\r')

pd.Series(observations).value_counts().plot.bar();
```

# Idealistic Scenario

- A man may meet 10,000 women in his life.
- A man will meet each woman up to 365 times, allowing for a full year of interactions, before he decides where that woman is located on the hot crazy matrix.

This is an idealistic scenario, offering extensive opportunities for observation and assessment, enabling a more informed decision-making process.

```python
observations = []
for i in range(10000):
    n = 365
    hot_vector = np.random.normal(loc=5, scale=3, size = n )
    crazy_vector = np.random.normal(loc=6, scale=2, size = n )
    
    area = plot_crazy_hot_matrix(hot_vector, crazy_vector, return_area=True, plot=False)
    observations.append(area)
    print(i, end='\r')

pd.Series(observations).value_counts().plot.bar();
```
